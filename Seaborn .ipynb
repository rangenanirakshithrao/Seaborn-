```python
# Loading a dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("Cars2015.csv")
```

```python
data
```

```python
# Line plot
sns.lineplot(x="Make",y="Type",data=data)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.lineplot(x="Length",y="Width",data=data,hue="Size")
```

```python
# Scatter Plot
sns.set(rc={'figure.figsize':(20,15)})
sns.scatterplot(x="Size",y="Make",data=data,hue="Drive")
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.scatterplot(x="CityMPG",y="HwyMPG",data=data,hue="FuelCap")
```

```python
# barplot
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x="CityMPG",y="HwyMPG",data=data,hue="Type")
```

```python
# count plot
sns.set(rc={'figure.figsize':(20,10)})
sns.countplot(x="Make",data=data)
```

```python
# distribution plot
sns.distplot(data["UTurn"])
```

```python
sns.distplot(data["UTurn"],bins=50,kde=False)
```

```python
# pair plot
sns.pairplot(data)
```

```python
# joint plot
sns.jointplot(x="CityMPG",y="HwyMPG",data=data,hue="Type")
```

```python
# box plot
sns.boxplot(x="Length",y="Width",data=data,hue="Drive")
```

```python
# correlation
correlation=data.corr()
sns.heatmap(correlation)
```
